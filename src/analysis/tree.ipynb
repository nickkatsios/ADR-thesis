```python
import pandas as pd
from graphviz import Digraph

# Load the updated CSV file
csv_file_path = './topic_info.csv'  # Replace with your file path
latest_topic_df = pd.read_csv(csv_file_path)

# Function to create the tree structure from the dataframe
def create_tree_structure(df):
    topic_tree = {
        "Module API Evolution Decision": {
            "Email Delivery Architecture": df.loc[df["Name"].str.contains("Email Delivery Architecture"), "Count"].values[0],
            "Data Abstraction Modeling Strategy": {
                "Data Management System Optimization": df.loc[df["Name"].str.contains("Data Management System Optimization"), "Count"].values[0],
                "Telemetry Metric Collection Strategy": {
                    "Messaging Telemetry Collection System": {
                        "Modular System Architecture": {
                            "Messaging System Design Approach": df.loc[df["Name"].str.contains("Messaging System Design Approach"), "Count"].values[0],
                            "Module Communication Decisions": df.loc[df["Name"].str.contains("Module Communication Decisions"), "Count"].values[0],
                        },
                        "Data Modeling Decisions": {
                            "Data Transformation Strategy": {
                                "Data Storage Decisions": df.loc[df["Name"].str.contains("Data Storage Decisions"), "Count"].values[0],
                                "Access Control Decisions": {
                                    "Data Schema Evolution Strategy": {
                                        "API Documentation Strategy": df.loc[df["Name"].str.contains("API Documentation Strategy"), "Count"].values[0],
                                        "Data Structure Transformation Proposal": df.loc[df["Name"].str.contains("Data Structure Transformation Proposal"), "Count"].values[0],
                                    },
                                    "User Authentication Architecture": df.loc[df["Name"].str.contains("User Authentication Architecture"), "Count"].values[0],
                                },
                            },
                            "Monitoring And Metric Management": df.loc[df["Name"].str.contains("Monitoring and Metric Management"), "Count"].values[0],
                        },
                    },
                    "Search Indexing Architecture Decision": df.loc[df["Name"].str.contains("Search Indexing Architecture Decision"), "Count"].values[0],
                },
            },
        },
        "Deployment Configurations Analysis": {
            "Cloud Infrastructure Architecture": {
                "Containerized Application Development": df.loc[df["Name"].str.contains("Containerized Application Development"), "Count"].values[0],
                "Infrastructure Decisions": {
                    "Cloud-Native Architecture Decisions": {
                        "Cluster Traffic Management Strategy": df.loc[df["Name"].str.contains("Cluster Traffic Management Strategy"), "Count"].values[0],
                        "Cloud Infrastructure Management": df.loc[df["Name"].str.contains("Cloud Infrastructure Management"), "Count"].values[0],
                    },
                    "Infrastructure Management Strategy": df.loc[df["Name"].str.contains("Infrastructure Management Strategy"), "Count"].values[0],
                },
            },
            "Modernizing Legacy Code Base": {
                "Software Component Decisions": {
                    "Software Development Architectural Decisions": df.loc[df["Name"].str.contains("Software Development Architectural Decisions"), "Count"].values[0],
                    "Software Architecture Decisions": {
                        "Test Strategy Evolution": df.loc[df["Name"].str.contains("Test Strategy Evolution"), "Count"].values[0],
                        "Code Architecture Decisions": {
                            "Dependency Management Decision": {
                                "Component Library Implementation Strategy": {
                                    "Modular Architecture Decisions": df.loc[df["Name"].str.contains("Modular Architecture Decisions"), "Count"].values[0],
                                    "Component Architecture Decisions": {
                                        "Library Component Distribution": df.loc[df["Name"].str.contains("Library Component Distribution"), "Count"].values[0],
                                        "Build System Configuration Decisions": df.loc[df["Name"].str.contains("Build System Configuration Decisions"), "Count"].values[0],
                                    },
                                },
                                "Release Process Standardization": df.loc[df["Name"].str.contains("Release Process Standardization"), "Count"].values[0],
                            },
                            "Architecture Decision Records (ADRs)": df.loc[df["Name"].str.contains("Architecture Decision Records (ADRs)"), "Count"].values[0],
                        },
                    },
                },
                "Code Formatting Decision Strategy": df.loc[df["Name"].str.contains("Code Formatting Decision Strategy"), "Count"].values[0],
            },
        },
    }
    return topic_tree

# Create the topic tree based on the dataframe
topic_tree = create_tree_structure(latest_topic_df)

# Function to add edges to the graph
def add_edges_tree(dot, parent, subtree):
    if isinstance(subtree, dict):
        for child, grandchild in subtree.items():
            dot.node(child, child, shape='ellipse', style='filled', fillcolor='lightblue', fontsize='18', width='2', height='2')
            dot.edge(parent, child)
            add_edges_tree(dot, child, grandchild)
    else:
        dot.node(str(subtree), f'{parent} ── ADRs: {subtree}', shape='ellipse', style='filled', fillcolor='lightblue', fontsize='18', width='2', height='2')
        dot.edge(parent, str(subtree))

# Initialize the Digraph with adjusted node style
dot = Digraph(comment='ADR Topic Tree')

# Adding the root nodes
for parent, subtree in topic_tree.items():
    dot.node(parent, parent, shape='ellipse', style='filled', fillcolor='lightblue', fontsize='18', width='2', height='2')
    add_edges_tree(dot, parent, subtree)

# Render the graph
dot.attr(size='10,10')
dot.attr(rankdir='TB')
dot.render('adr_topic_tree', format='pdf')
```
